```python
#Choosing the y column(Target parameter)
output_var = pd.DataFrame(df['Close'])
feature = ['Open','High','Low','Volume']
                    
```

```python
output_var.shape
plt.plot(output_var)
```

```python
#LSTM are sensitive to scale of data,WE scale it bt 0 to 1
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
#or df1=scaler.fit_transform(df1)
output_var =scaler.fit_transform(np.array(output_var).reshape(-1,1))
feature_transform = scaler.fit_transform(df[feature])
```

```python
pd.DataFrame(feature_transform).head()
```

```python
#Splitting train and test data
#this is a time series data so we divide it w.r.t date,Here next data is dependent on previous data
from sklearn.model_selection import TimeSeriesSplit
timesplit= TimeSeriesSplit(n_splits=10)
for train_index, test_index in timesplit.split(feature_transform):
        X_train, X_test = feature_transform[:len(train_index)], feature_transform[len(train_index): (len(train_index)+len(test_index))]
        y_train, y_test = output_var[:len(train_index)].ravel(), output_var[len(train_index): (len(train_index)+len(test_index))].ravel()
```

```python
pd.DataFrame(X_train)
```

```python
pd.DataFrame(y_test)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
from tensorflow.keras.models import load_model
```

```python
#LSTM needs data in 3D format. Thus we tranform the training and test set.
trainX =np.array(X_train)
testX =np.array(X_test)
X_train = trainX.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = testX.reshape(X_test.shape[0], 1, X_test.shape[1])
```

```python

trainX.shape[1]
```

```python
lstm = Sequential()
lstm.add(LSTM(32, input_shape=(1,4),activation='relu',return_sequences=False))
lstm.add(Dense(1))
```

```python
lstm.compile(loss='mean_squared_error',optimizer='adam')
```

```python
history=lstm.fit(X_train, y_train, epochs=100, batch_size=8, verbose=1, shuffle=False)
```

```python
y_pred = lstm.predict(X_test)
```

```python
lstm.summary()
```

```python
plt.plot(y_test, label='True Value')
plt.plot(y_pred, label='LSTM value')
plt.title('Prediction by LSTM')
plt.xlabel('Time Scale')
plt.ylabel('Scaled ')
plt.legend()
plt.show()
```
